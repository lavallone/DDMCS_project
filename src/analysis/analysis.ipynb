### üìä Plot Data

#### Cumulative number of unique posts üìÉ

```python
from plots import plot_daily_posts
import matplotlib.pyplot as plt
import pandas as pd
import os

daily_post_counts_file = os.path.join(f"../../data/quantitative_analysis", f'daily_post_counts.csv')
line_data = pd.read_csv(daily_post_counts_file)

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(30, 10))
fig.subplots_adjust(wspace=0.3, bottom=0.15)
for idx,topic in enumerate(line_data["topic"].unique()):
    line_topic_data = line_data[line_data["topic"] == topic]
    plot_daily_posts(line_data=line_topic_data, ax=axs[idx], topic=topic)
```

#### Number of interactions ü´Ç

```python
from plots import plot_interaction_dist
import matplotlib.pyplot as plt
import pandas as pd
import os

interaction_dist_file = os.path.join(f"../../data/quantitative_analysis", f'interaction_dist.csv')
scatter_data = pd.read_csv(interaction_dist_file)

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(30, 10))
fig.subplots_adjust(wspace=0.3, bottom=0.15)
for idx,topic in enumerate(scatter_data["topic"].unique()):
    scatter_topic_data = scatter_data[scatter_data["topic"] == topic]
    plot_interaction_dist(scatter_data=scatter_topic_data, ax=axs[idx], topic=topic)
```
